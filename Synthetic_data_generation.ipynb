```python
import csv
import pandas as pd
import numpy as np
import random
```

```python
filename = '100-nodes.csv'
```

```python
df = pd.read_csv(filename)
```

```python
df
```

```python
for i in range(1900):
    df = df.append({
        'CPU-Utilization': random.uniform(0, 100),
        'Memory-Usage': random.uniform(0, 100),
    }, ignore_index=True)
```

```python
df.to_csv('2000-nodes.csv', index=False)
```
